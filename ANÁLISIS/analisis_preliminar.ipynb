```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Configuración estética para los gráficos del TIF
sns.set_theme(style="whitegrid")

# Cargamos el dataset limpio que ya guardamos antes
df = pd.read_csv('../DATOS/Datos procesados/dataset_final_vilo.csv', sep=';')

# ¡Listo para empezar a graficar sin tocar nada de la limpieza!
print(f"Dataset cargado con {len(df)} registros.")
```

```python
# 1. Calculamos las medidas de tendencia central
media = df['votos'].mean()
mediana = df['votos'].median()
moda = df['votos'].mode()[0] # .mode() devuelve una serie, tomamos el primer valor
desviacion_est = df['votos'].std()
minimo = df['votos'].min()
maximo = df['votos'].max()

print(" Estadísticas descriptivas de los Votos ---")
print(f"Media (Promedio): {media:.2f}")
print(f"Mediana (Valor central): {mediana:.2f}")
print(f"Moda (Valor más frecuente): {moda}")
print(f"Desviación estándar: {desviacion_est:.2f}")
print(f"Rango: de {minimo} a {maximo} votos")

# 2. Análisis por resultado (Aprobado vs Desaprobado)
stats_por_resultado = df.groupby('ganador_bin')['votos'].agg(['mean', 'median', 'std', 'count'])
print("\n Estadísticas por Resultado (0: Desaprobado, 1: Aprobado) ---")
print(stats_por_resultado)
```
